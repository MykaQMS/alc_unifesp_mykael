## Exercício 8

Gere 100 matrizes aleatórias $5 \times 5$ com entradas inteiras entre $-5$ e $5$.

Para cada matriz:

1. Calcule o posto.
2. Verifique se a matriz é invertível.
3. Registre a frequência de cada valor de posto.
4. Apresente: uma tabela de frequências.

**Sugestão:**

```python
np.random.randint(-5, 5, (5, 5))

````python
# Criando arrays para armazenar os resultados

postos = []
matrizes_intertiveis = []

# Gerando as 100 matrizes aleatórias 5x5 (valores entre -5 e 5)
#usando (-5,6,...) como argumento pois a função randint() não considera o limite superior na criação da matriz

# Loop principal
for i in range(100):
  matriz = np.random.randint(-5, 6, (5, 5))
  posto = matrix_rank(matriz)
  postos.append(posto)
  if posto == 5:
    matrizes_intertiveis.append(matriz)
  else:
    pass

  print(f"\n-> Posto da matriz: {posto}\n")
  if posto == 5:
    print("-> A matriz é invertível\n")
  else:
    print("-> A matriz não é invertível\n")
  print(matriz)
````

````python
# contar registros

print(f"Total de matrizes geradas: {len(postos)}")
print(f"Quantidade de matrizes intertiveis: {len(matrizes_intertiveis)}")
print(f"Quantidade de matrizes não intertiveis: {len(postos) - len(matrizes_intertiveis)}")
````

````python
# Verificando a frequência de cada posto

import pandas as pd

freq_postos = pd.Series(postos).value_counts().sort_index()
print(freq_postos)
````

````python
tabela_freq = freq_postos.reset_index()
tabela_freq.columns = ['Posto', 'Frequência']
print("Tabela de frequências por posto:")
tabela_freq
````

# **Comentários sobre o exercício**

---

1. Observamos que, pela aleatoriedade da geração das matrizes, a chance de termos linhas ou colunas linearmente dependentes é muito baixa. Portanto, ocorreu que obtivemos posto 5 para todas as 100 matrizes geradas.
2. Observamos também que, por consequência do supracitado, todas as matrizes foram invertíveis.

## Exercício 9 — Espaço Nulo e Soluções de Sistemas Lineares

Considere o sistema:

$$
Ax = b
$$

com

$$
A =
\begin{bmatrix}
1 & 2 & 1 \\
2 & 4 & 2 \\
1 & 1 & 0
\end{bmatrix}
\quad \text{e} \quad
b =
\begin{bmatrix}
1 \\
2 \\
1
\end{bmatrix}
$$

1. Calcule o espaço nulo de $A$.
2. Resolva o sistema.

Verifique se o sistema possui:

- solução única  
- infinitas soluções  
- nenhuma solução

````python
# Definindo as matrizes A e b para os cálculos seguintes
A = np.array([[1, 2, 1], [2, 4, 2], [1, 1, 0]])
b = np.array([[1], [2], [1]])

print("Matriz A:")
print(A)
print("\nMatriz b:")
print(b)
````

````python
#1. Calculando o espaço nulo de A
# Para calcular o espaço nulo de uma matriz A, podemos usar a função null_space do scipy.linalg
# que retorna uma base para o espaço nulo da matriz.
espaco_nulo_A = null_space(A)
print("\nEspaço nulo de A:")
print(espaco_nulo_A)
````

````python
#2. Resolvendo o sistema linear Ax = b
# Para resolver o sistema linear Ax = b, podemos usar a função solve do numpy.linalg
try:
    solucao = np.linalg.solve(A, b)
    print("\nSolução única do sistema Ax = b:")
    print(solucao)
except np.linalg.LinAlgError:
    print("\nO sistema Ax = b não tem solução única (pode ser indeterminado ou impossível).")
````

````python
# Descobrindo se o sistema é indeterminado ou impossível
# Para isso, podemos comparar o posto da matriz A com o posto da matriz aumentada [A | b]

matriz_aumentada = np.column_stack((A, b))

posto_A = np.linalg.matrix_rank(A)
posto_aumentada = np.linalg.matrix_rank(matriz_aumentada)
n_variaveis = A.shape[1]

print(f"\nPosto de A: {posto_A}")
print(f"Posto de [A | b]: {posto_aumentada}")

if posto_A != posto_aumentada:
    print("Sistema impossível (sem soluções).")
elif posto_A == posto_aumentada == n_variaveis:
    print("Sistema possível e determinado (solução única).")
else:
    print("Sistema possível e indeterminado (infinitas soluções).")
````

# **Comentários sobre o exercício**

---

1. Observamos que o sistema possui solução, mas não é única.
2. Pela regra do posto de A e da matriz aumentada de [A | b] temos que o sistema possui soluções infinitas (posto de A = posto de aug(A) = n).

## Exercício 10

Gere dois vetores aleatórios em $\mathbb{R}^{10}$.

Calcule a distância entre eles usando:

- norma 1  
- norma 2  
- norma infinito  

Repita o experimento 100 vezes e compare os resultados médios.

Apresente:

- média das distâncias  
- desvio padrão  

Discuta: qual norma tende a produzir valores maiores?

````python
# Gerando os arrays vazios para armazenar os resultados

nrm_1 = []
nrm_2 = []
nrm_inf = []

# Loop principal de iteração

for i in range(100):

  # Gerando os dois vetores aleatórios em R^10

  x = np.random.rand(10)
  y = np.random.rand(10)

  dist = x - y

  norma_1 = np.linalg.norm(dist, ord=1)
  norma_2 = np.linalg.norm(dist, ord=2)
  norma_infinito = np.linalg.norm(dist, ord=np.inf)
  print(f"\nNorma 1: {norma_1}")
  print(f"Norma 2: {norma_2}")
  print(f"Norma infinito: {norma_infinito}")

  nrm_1.append(norma_1)
  nrm_2.append(norma_2)
  nrm_inf.append(norma_infinito)

# Calculando a média e desvio padrão das distâncias (normas)

media_nrm_1 = np.mean(nrm_1)
media_nrm_2 = np.mean(nrm_2)
media_nrm_inf = np.mean(nrm_inf)

sd_nrm_1 = np.std(nrm_1)
sd_nrm_2 = np.std(nrm_2)
sd_nrm_inf = np.std(nrm_inf)

print(f"\nMédia das distâncias (norma 1): {media_nrm_1}")
print(f"Média das distâncias (norma 2): {media_nrm_2}")
print(f"Média das distâncias (norma infinito): {media_nrm_inf}")
print(f"\nDesvio padrão das distâncias (norma 1): {sd_nrm_1}")
print(f"Desvio padrão das distâncias (norma 2): {sd_nrm_2}")
print(f"Desvio padrão das distâncias (norma infinito): {sd_nrm_inf}")
````

# **Comentários sobre o exercício**

---

Observamos pelos resultados de média e desvio de cada norma que a que tende a produzir valores mais altos é a **norma 1**. Isso porque ela não usa raiz quadrada no cálculo, então o resultado tende a ser maior.

## Exercício 11

Investigue empiricamente a seguinte questão:

**Qual é a probabilidade de uma matriz aleatória $n \times n$ ser singular?**

### Procedimento:

- Para $n = 2, 3, 4, 5, 6$
- Gere 1000 matrizes aleatórias
- Calcule quantas são singulares

### Apresente:

- gráfico: probabilidade × dimensão

### Discuta:

- o comportamento observado

````python
# Determinando as dimensões n da matriz aleatória

dim = [2,3,4,5,6]
````

````python
# Criando container para armazenar os resultados

probabilidades = []

# Loop de iteração

for n in dim:
  singulares = 0

  for i in range(1000):
    A = np.random.randint(-5,6, (n,n))
    if np.linalg.matrix_rank(A) < n:
      singulares += 1
    else:
      pass
  prob = singulares / 1000
  probabilidades.append(prob)

  print(f"\nDimensão da matriz: {n}")
  print(f"Número de matrizes singulares: {singulares}")
  print(f"Probabilidade de uma matriz ser singular: {prob}")
````

````python
resultado = pd.DataFrame({
    'Dimensão': dim,
    'Probabilidade': probabilidades
})

print(resultado)
````

````python
# Criando o gráfico probabilidade x dimensão

import matplotlib.pyplot as plt

plt.plot(dim, probabilidades, marker='o')
plt.xlabel('Dimensão')
plt.ylabel('Probabilidade')
plt.title('Probabilidade de matriz aleatória ser singular')
plt.grid(True)
plt.show()
````

# **Comentários sobre o exercício**

---

1. Observamos que, ao aumentar a dimensão n da matriz, a probabilidade de ser singular diminui.
2. Pelo gráfico, observamos uma tendência de estabilização, algo como a se a probabilidade tendesse para 0 conforme aumentamos a dimensão da matriz aleatória.
