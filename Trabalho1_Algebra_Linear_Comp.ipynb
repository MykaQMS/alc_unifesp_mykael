## Exercício 9 — Espaço Nulo e Soluções de Sistemas Lineares

Considere o sistema:

$$
Ax = b
$$

com

$$
A =
\begin{bmatrix}
1 & 2 & 1 \\
2 & 4 & 2 \\
1 & 1 & 0
\end{bmatrix}
\quad \text{e} \quad
b =
\begin{bmatrix}
1 \\
2 \\
1
\end{bmatrix}
$$

1. Calcule o espaço nulo de $A$.
2. Resolva o sistema.

Verifique se o sistema possui:

- solução única  
- infinitas soluções  
- nenhuma solução

````python
# Definindo as matrizes A e b para os cálculos seguintes
A = np.array([[1, 2, 1], [2, 4, 2], [1, 1, 0]])
b = np.array([[1], [2], [1]])

print("Matriz A:")
print(A)
print("\nMatriz b:")
print(b)
````

````python
#1. Calculando o espaço nulo de A
# Para calcular o espaço nulo de uma matriz A, podemos usar a função null_space do scipy.linalg
# que retorna uma base para o espaço nulo da matriz.
espaco_nulo_A = null_space(A)
print("\nEspaço nulo de A:")
print(espaco_nulo_A)
````

````python
#2. Resolvendo o sistema linear Ax = b
# Para resolver o sistema linear Ax = b, podemos usar a função solve do numpy.linalg
try:
    solucao = np.linalg.solve(A, b)
    print("\nSolução única do sistema Ax = b:")
    print(solucao)
except np.linalg.LinAlgError:
    print("\nO sistema Ax = b não tem solução única (pode ser indeterminado ou impossível).")
````

````python
# Descobrindo se o sistema é indeterminado ou impossível
# Para isso, podemos comparar o posto da matriz A com o posto da matriz aumentada [A | b]
matriz_aumentada = np.hstack((A, b))
posto_A = matrix_rank(A)
posto_aumentada = matrix_rank(matriz_aumentada)
print(f"\nPosto de A: {posto_A}")
print(f"Posto da matriz aumentada [A | b]: {posto_aumentada}")
if posto_A < posto_aumentada:
    print("O sistema é impossível (sem soluções).")
elif posto_A == posto_aumentada:
    print("O sistema é possível (tem soluções).")
else:
    print("O sistema é indeterminado (tem infinitas soluções).")
````
